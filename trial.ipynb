```python
import torch
import torch.nn as nn

# Create a BatchNorm2d layer for 16 channels
batch_norm = nn.BatchNorm2d(num_features=16)

# Example input tensor with shape (N, C, H, W)
input_tensor = torch.randn(8, 16, 32, 32)  # Batch size of 8, 16 channels, 32x32 image

# Apply BatchNorm2d
output_tensor = batch_norm(input_tensor)

print("Input shape:", input_tensor.shape)
print("Output shape:", output_tensor.shape)
```
